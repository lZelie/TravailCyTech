## 1. Différences majeures entre les réseaux de neurones standards et les réseaux de neurones convolutifs (ConvNet)  :

**1. Architecture :**

* **Réseaux de neurones standards :** architecture simple avec des couches entièrement connectées. Chaque neurone est connecté à tous les neurones de la couche suivante.
* **ConvNet :** architecture plus complexe avec des couches convolutives et de pooling. Les couches convolutives exploitent la structure spatiale des données, tandis que les couches de pooling réduisent la dimensionnalité des données.

**2. Fonctionnement :**

* **Réseaux de neurones standards :** traitent les données d'entrée comme un vecteur plat. Apprennent des relations statistiques entre les caractéristiques individuelles.
* **ConvNet :** traitent les données d'entrée comme une grille (images) ou une séquence (texte). Apprennent des motifs locaux et des relations spatiales entre les caractéristiques.

**3. Applications :**

* **Réseaux de neurones standards :** large éventail d'applications, y compris la classification, la régression et le clustering.
* **ConvNet :** particulièrement adaptés aux tâches de vision par ordinateur, telles que la reconnaissance d'images, la détection d'objets et la segmentation d'images.

## 2. Recuit simulé :

**1. Impact du paramètre de décroissance α dans une décroissance de température géométrique  :**

* **Valeur de α élevée :** 
    * Refroidissement rapide.
    * Moins d'exploration de l'espace de recherche.
    * Convergence plus rapide vers un optimum local.
    * Risque accru de manquer l'optimum global.

* **Valeur de α faible :** 
    * Refroidissement lent.
    * Plus d'exploration de l'espace de recherche.
    * Meilleure chance de trouver l'optimum global.
    * Convergence plus lente.

**2. Évaluation de la différence de score et critère de Metropolis  :**

**Différence de score :**

* Calculée entre le score du candidat (solution proposée) et le score de son voisin (solution actuelle).
* Détermine si le candidat est une meilleure solution que le voisin.

**Critère de Metropolis :**

* Détermine si le candidat est accepté ou non comme nouvelle solution.
* Si la différence de score est positive, le candidat est toujours accepté.
* Si la différence de score est négative, le candidat est accepté avec une probabilité décroissant avec la différence de score et la température.

**3. Convergence de l'algorithme du recuit simulé w  :**

* La température décroît progressivement, limitant l'exploration de l'espace de recherche.
* L'algorithme se concentre sur les zones prometteuses de l'espace de recherche.
* La probabilité d'accepter des solutions de moindre qualité diminue, favorisant la convergence vers un optimum.

## Apprentissage par renforcement - Modélisation du problème

1. **Définition de l'état :**

    L'état de l'agent peut être défini par sa position sur la grille (x, y).

2. **Définition de l'action :**

    L'agent peut choisir parmi 4 actions possibles :

    * Haut
    * Bas
    * Gauche
    * Droite

3. **Définition de la récompense :**

    * +10 : Atteindre la case verte (objectif)
    * -1 : Se heurter à un obstacle
    * -0.1 : Se déplacer sur une case vide

4. **Définition de la fonction de transition :**

    La fonction de transition détermine la probabilité de passer d'un état à un autre en fonction de l'action choisie.

5. **Définition de la métrique d'évaluation :**

    La métrique d'évaluation est la somme des récompenses cumulées sur un certain nombre d'épisodes.

6. **Modélisation mathématique :**

    L'environnement peut être modélisé par un processus de décision markovien (MDP) discret.

7. **Choix de l'algorithme d'apprentissage :**

    Plusieurs algorithmes d'apprentissage par renforcement peuvent être utilisés pour résoudre ce problème, tels que Q-Learning ou Monte Carlo Tree Search.

## Stratégie

1. **Définition  :**

    Une stratégie est une fonction qui mappe chaque état à une action.

2. **Exemple  :**

    Une stratégie simple pourrait être de toujours se déplacer vers la droite jusqu'à atteindre la case verte.

## Stratégie optimale

1. **Définition  :**

    Une stratégie optimale est une stratégie qui maximise la métrique d'évaluation.

2. **Détermination  :**

    La stratégie optimale peut être trouvée par des algorithmes d'apprentissage par renforcement.

## Deviner la stratégie optimale

1. **Analyse du problème  :**

    Dans ce cas simple, il est possible de deviner la stratégie optimale.

2. **Justification :**

    * La case verte se trouve en bas à droite.
    * Il n'y a pas d'obstacles entre l'agent et la case verte.
    * Se déplacer vers la droite est la meilleure option pour se rapprocher de la case verte.

3. **Stratégie optimale  :**

    * S00, S01, S20, S21 → ↓
    * S02, S12, S13, S01, S32 → →
    * S03 → ↑
    * S30 → ← 

## Algorithme pour trouver la stratégie optimale

1. **Q-Learning :**

    Q-Learning est un algorithme d'apprentissage par renforcement qui permet à l'agent d'apprendre une stratégie optimale par essais et erreurs.

2. **Fonctionnement  :**

    L'algorithme Q-Learning utilise une fonction de valeur d'action Q(s, a) pour estimer la valeur attendue de chaque action dans chaque état.

3. **Mise à jour de la fonction Q  :**

    La fonction Q est mise à jour à chaque interaction avec l'environnement en fonction de la récompense reçue et de la valeur maximale de la fonction Q pour l'état suivant.

4. **Convergence :**

    L'algorithme Q-Learning converge vers une stratégie optimale après un nombre suffisant d'interactions avec l'environnement.

## Conclusion

L'apprentissage par renforcement est une approche puissante pour résoudre des problèmes de planification et de décision. Dans ce cas simple, il est possible de deviner la stratégie optimale. Cependant, pour des problèmes plus complexes, des algorithmes d'apprentissage par renforcement sont nécessaires pour trouver la stratégie optimale.

# Exercice IV : Recuit simulé
Dans cet exercice, nous souhaitons résoudre par l’algorithme du recuit simulé une version très simplifié du problème d’affectation de places de parkings. Nous définissons le problème comme suit :
* Nous avons N véhicules et `k` places de parkings. Nous supposons que `k` est strictement supérieure à `N`.
* Chaque place de parking `Pi` est connue par son adresse que nous exprimons par un couple de coordonnées `(xi, yi)`.
* Chaque véhicule `Vj` est aussi connu par une adresse (par exemple l’adresse du logement ou du lieu de travail du propriétaire). Notons `(zj, tj)` les coordonnées définissant une telle adresse. 

Notre objectif est d’affecter à chaque véhicule une place de parking aussi proche que possible de l’adresse qui lui est associée. On appelle affectation un ensemble de N couples `{(Pi, Vj)}` dans lequel tous les `Pi`, ainsi que tous les `Vj`, sont deux à deux différents.
1. Donnez le nombre total d’affectations. En déduire qu’il est utile d’utiliser le recuit simulé pour la résolution du problème.
2. Donnez une représentation des affectations, une définition du voisinage et la fonction à minimiser.

## Réponses à l'exercice sur le recuit simulé pour l'affectation de places de parking :

1. **Nombre total d'affectations et justification du recuit simulé :**

    **Nombre d'affectations :**
    
    Le nombre total d'affectations possibles est donné par le nombre de combinaisons possibles de `N` éléments parmi `k` éléments, où `k` est supérieur à `N`.
    
    Calculons le nombre d'arrangements possibles de `N` éléments pris parmi `k` :
    
    `A_N^k = \frac{k!}{(k-N)!}`
    
    Cependant, il y a des permutations identiques (correspondant à la même affectation) dans cet arrangement. Il faut donc diviser par le nombre de permutations de `N` éléments :
    
    `N!`
    
    Donc, le nombre total d'affectations distinctes est :
    
    `C_N^k = \frac{k!}{(k-N)! \times N!}`
    
    **Justification du recuit simulé :**
    
    Le nombre d'affectations possibles est exponentiel en `N`. Pour des valeurs raisonnables de `N` et `k`, il est impossible de tester toutes les affectations. Le recuit simulé est une méthode probabiliste qui permet d'explorer l'espace des solutions et de trouver une solution de bonne qualité, sans nécessairement garantir l'optimalité.

2. **Représentation des affectations, voisinage et fonction à minimiser :**
    
    **Représentation des affectations :**
    
    Une affectation peut être représentée par un tableau de `N` éléments, où chaque élément est un indice de place de parking. Par exemple, l'affectation `[2, 1, 0]` signifie que le véhicule `V1` est affecté à la place `P2`, le véhicule `V2` à la place `P1` et le véhicule `V3` à la place `P0`.
    
    **Définition du voisinage :**
    
    Un voisinage d'une affectation `A` peut être défini par l'ensemble des affectations qui peuvent être obtenues en échangeant deux places de parking dans `A`. Par exemple, si `A` est `[2, 1, 0]`, un voisin de `A` est `[2, 0, 1]`.
    
    **Fonction à minimiser :**
    
    La fonction à minimiser est la distance totale parcourue par tous les véhicules pour se rendre à leur place de parking attribuée. La distance peut être calculée en utilisant la distance euclidienne entre les coordonnées de la place de parking et les coordonnées du véhicule.
    
    **Exemple de fonction à minimiser :**
    
    `f(A) = Σ_i^N d((xi, yi), (zj, tj))`
    
    où `d` est la distance euclidienne.

**Remarques :**

* D'autres définitions de voisinage et de fonction à minimiser peuvent être utilisées.
* Le choix de la fonction de décroissance de la température est crucial pour la performance de l'algorithme.
* L'algorithme du recuit simulé est un outil puissant pour résoudre des problèmes d'optimisation complexes.
